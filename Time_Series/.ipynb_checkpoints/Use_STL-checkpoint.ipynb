# Descomposicion de Series de Tiempo

La descomposicion de series de tiempo es una herramienta analitica esencial para entender nuestro datos y a si poder mejorar los pronosticos.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline

# the main library has a small set of functionality
from stldecompose import decompose, forecast
from stldecompose.forecast_funcs import (naive,
                                         drift, 
                                         mean, 
                                         seasonal_naive)


%load_ext autoreload
%autoreload 2
```

Usaremos los datos de mi trabajo de grado de BS que tiene algunos datos vacios

```python
data= pd.read_excel('Data_Categoria7.xlsx',encoding='ISO-8859-1')
print(type(data))
data.head()

```

Como STLDecompose se basa en funcionalidades `statsmodels`, entonces `STLDecompose` requiere que sus conjuntos de datos cumplan con los siguientes requisitos:
1. Observaciones continuas (Sin vacios )
2. Un indice de tiempo de  `pandas` `DateTimeIndex`

Como ambas caracteriscas de depende del contexto del problema `STLDecompose` no ajusta del conjunto de datos si no que deja que el usurio aplique las herramientas que considere para estas Tareas, el paquete que se usa para estas tareas es pandas.

```python
# Creacion de rango de fechas Index con frecuencia mensual
date_range = pd.date_range('1/1/2008',periods=60,freq='M')

# Transformadon el rango en un dat frame para poder manipularlo
df_dates = pd.DataFrame(date_range, columns=['date'])

#a√±adiendo en data frame de fechas al dataframe principal

data['date'] = df_dates

#reconvirtiendo las fechas en un tipo de dato fecha para manipularlo como serie de tiempo

data['datetime'] = pd.to_datetime(data['date'])

# Configurando el campo datetime como index

data = data.set_index(['datetime'])

#imprimiendo el resultado

data.head()



```

```python
# configruando la frecuencia

data = data.asfreq(freq='M')
data.index

```

```python
data.drop(['date'], axis=1, inplace=True)
data.drop(['Fecha'], axis=1,inplace=True)
```

```python
data.plot()
```

# Descomposicion

El metodo asociado a la fucion de descomposisicion requiere de un marco observacion y el periodo de la estacionalidad. Este `periodo` es en terminanos de la posiciones de indice , por lo es necesario  que usuario tenga precaucion


```python
stl = decompose(data, period=12)

stl
```

El objeto resultante es una version extendida de `statsmodels.tsa.seasonal.DecomposeResult`. Como el objeto de `statsmodels` , los valores de las matrices estan disponibles en el objecto (las Observaciones; y los componentes de Tendencia y estacionalidad). atributo extra (el ciclo estacional promedio) se ha a√±adido con el objetivo de pronosticar 

Heradamos el metodo `.plot()` incorporadp en el objeto on the object.

```python
stl.plot();
```

# Pronostico

Si bien la descomposicion nos muestra informacion interesante, `STLDecompose` tambien brinda algunas capacidades relativamente sencillas para usar la descomposicion para pronosticar. 

Usaremos el mismo conjunto de datos pero solo una parte para poder comparar nuestro pronostico

```python
len(data)
```

```python
short_data = data.head(48)
short_data.plot()
```

```python
# apply the decomp to the truncated observation
short_stl = decompose(short_data, period=12)

short_stl
```

El metodo `forecast()`  requiere los siguiente argumuentos:
- EL ajuste `DecomposeResult`
- El numero de periodos a pronosticar
- la funcion especifica que aplicaqueros a nuestro pronostico

hay varias funciones predefinidas que pueden ser importadas de el modulo `stldecompose.forecast_funcs`. esta implemenactacion esta basada [Hyndman's online textbook](https://www.otexts.org/fpp/2/3). el usuario tambien puede definir su propia fucion de pronosticos, siguiendo los patrones de lafuciones definidas 

El tipo retorno del metodo  `forecast()` es un `pandas.Dataframe` con una columna  que repsenta el pronostico y un apropiado  `DatetimeIndex`.

```python
fcast = forecast(short_stl, steps=12, fc_func=drift)

fcast.head()
```

If desired, we can then plot the corresponding components of the observation and forecast to check and verify the results.

```python
plt.plot(data, '--', label='truth')
plt.plot(short_data, '--', label='obs')
plt.plot(short_stl.trend, ':', label='stl.trend')
plt.plot(fcast, '-', label=fcast.columns[0])

plt.xlim('2008','2018'); plt.ylim(0,1000);
plt.legend();
```

To include the estimated seasonal component in the forecast, use the boolean `seasonal` keyword.

```python
fcast = forecast(short_stl, steps=12, fc_func=drift, seasonal=True)

plt.plot(data, '--', label='truth')
plt.plot(short_data, '--', label='obs')
plt.plot(short_stl.trend, ':', label='stl.trend')
plt.plot(fcast, '-', label=fcast.columns[0])

plt.xlim('2008','2013'); plt.ylim(0,1000);
plt.legend();
```
